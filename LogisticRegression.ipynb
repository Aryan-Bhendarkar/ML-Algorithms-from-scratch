### Implementation of Logistic Regression from scratch 😁

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Generate data for training and testing
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = (X>1).astype(int)

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Binary Classification")
plt.show()
```

```python
# Define Sigmoid function 
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Prediction function
def predict(w, b, X):
    z = np.dot(w.T, X) + b
    y_pred = sigmoid(z)
    return y_pred

# Compute the loss 
def compute_loss(y, y_pred):
    loss = -np.mean(y*np.log(y_pred) + (1-y)*np.log(1-y_pred))
    return loss 

# Gradient decent 
def compute_gradient(X, y, y_pred):
    n = len(y)
    dw = np.sum((y_pred-y)*X)/n
    db = np.sum(y_pred - y)/n
    return dw, db
```

```python
# Model 

w = np.zeros(X.shape[1], )
b = 0.0
learning_rate = 0.1
epoches = 1000
losses = []

for epoch in range(epoches):
    y_pred = predict(w, b, X)
    loss = compute_loss(y, y_pred)
    dw, db = compute_gradient(X, y, y_pred)
    w -= learning_rate*dw
    b -= learning_rate*db
    losses.append(loss)
    if(epoch%100 == 0):
        print(f"Epoch {epoch}/1000: Loss = {loss:.4f}")
    
print(f"Trained weight: {w}, bias: {b:.2f}")
```
