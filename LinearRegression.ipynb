### Implement Linear Regression from Scratch 😁


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Dataset 
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Data visualization 
plt.scatter(X, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Dataset")
plt.show()
```

##### Linear Regression implementation 

```python
# Parameters Initialization
w = 0.0
b = 0.0
learning_rate = 0.01
epoches = 1000

# Prediction 
def predict(X , w, b):
    y_pred = w*X + b
    return y_pred

# Compute Loss 
def compute_loss(y, y_pred):
    return np.mean((y_pred-y)**2)

# Gradient decent
def compute_gradient(X, y, y_pred):
    n = len(y)
    dw = (-2/n)*np.sum((y-y_pred)*X)
    db = (-2/n)*np.sum((y-y_pred))
    return dw, db

losses = []
for epoch in range(epoches):
    y_pred = predict(X, w, b)
    loss = compute_loss(y, y_pred)
    losses.append(loss)
    dw, db = compute_gradient(X, y, y_pred)
    w = w - learning_rate*dw
    b = b - learning_rate*db
    if epoch%100 == 0:
        print(f"Epoch {epoch}: Loss = {loss: .4f}")
        
print(f"Trained weight: {w:.2f}, bias: {b:.2f}")
```

```python
# Visualize the prediction v/s real values 
plt.scatter(X, y, label="Data points")
plt.plot(X, predict(X, w, b), color='red', label="Fitted line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss curve")
plt.show()
```
